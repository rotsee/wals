```python
# FIND THE MOST UNCOMMON PROPERTIES

lang = df[df["iso_code"] == CODE].fillna("")

result = []
for c in columns:
    if not c in df.columns:
        continue
    if not lang[c].iloc[0]:
        continue
    val = lang[c].iloc[0]
    result.append({
        "key": c,
        "value": val,
        "percent": round(counts[c][val] / counts[c].sum() * 100),
    })
res = pd.DataFrame(result)
res.sort_values("percent").head(10)
```
